# Data Visualization Basics
## Using a csv file to extract and visualize data

```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from numpy.random import randn
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

## first let's create a line chart from a list object

```python
x = range(1,10)
y = [2, 1, 5, 3, 12, 1, 15, 5, 6]
plt.plot(x,y)
```

## Now let's create charts from the csv file
### Creating line chart from pandas object

```python
address = "mtcars.csv"
cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
mpg = cars['mpg']
mpg.plot()
```

```python
df = cars[['mpg', 'cyl', 'wt']]
df.plot()
```

## creating bar charts

```python
plt.bar(x,y)
```

```python
mpg.plot(kind='bar')
```

```python
# horizontal bar chart
mpg.plot(kind='barh')
```

```python
df = cars[['mpg', 'cyl', 'wt']].head()
df.plot(kind='bar')
```
